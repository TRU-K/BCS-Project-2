```python
# Import libraries
import talib as ta
from talib import abstract
import pandas as pd
```

```python
# Read in data from csv
df = pd.read_csv('../Data/ETF_top_8_data.csv', header=[0, 1], parse_dates=True, index_col=[0])
```

```python
# All availible functions within TA-Lib
ta.get_function_groups()
```

```python
def bbands_create_df(timeperiod, nbdevup, nbdevdn, ticker):
    data=df[ticker]['close']
    bbands_abstract = ta.BBANDS(data, 20, 2, 2)
    BB_df = pd.DataFrame(bbands_abstract).transpose()
    BB_df.rename(columns={0:'BB_UPPER', 1:'BB_MIDDLE', 2:'BB_LOWER'}, inplace=True)
    return BB_df
```

```python
ticker_list = list(dict.fromkeys((df.droplevel(axis=1, level=[-1]))))
```

```python
bbands_df = pd.DataFrame()
for x in ticker_list:
    bbands = bbands_create_df(timeperiod=20, nbdevup=5, nbdevdn=5, ticker=x)
    df_temp = pd.concat({f"{x}": bbands}, axis=1)
    for BB in bbands.columns:    
        df[f'{x}', f'{BB}'] = bbands[BB]
df = df.sort_index(axis=1)
```
