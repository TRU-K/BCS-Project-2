```python
XLE_df = df_ticker_data['XLE'].dropna().copy()
XLE_df.tail(100)
```

# New Method Bernoulli Classifier

# The classifier is one of 3 in Naive Bayes library. It binarises the data (here we binarised sell / buy signalls).


```python
from sklearn.model_selection import train_test_split
```

```python
# Create our features
X = XLE_df.drop(columns='signal')

# Create our target
y = XLE_df['signal']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, test_size = 0.30)
```

```python
from sklearn.naive_bayes import BernoulliNB
BernNB = BernoulliNB(binarize = True)
BernNB.fit(X_train, y_train)
y_pred=BernNB.predict(X_test)
print(y_pred)
```

# Model Evaluation

```python
from sklearn import metrics
# Accuracy score
print(metrics.accuracy_score(y_test, y_pred))
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

by = sb.distplot(y_pred, 
                kde = True, 
               color = 'green', 
               hist_kws={'linewidth':30, 'alpha':1})

by.set(xlabel='Bernoulli', ylabel = 'Frequency', title = 'Bernoulli Distribution')
plt.show()
```

```python
from sklearn.metrics import classification_report
print (classification_report(y_test, y_pred, target_names = ('buy', 'sell')))
```
